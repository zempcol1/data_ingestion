#### <b>Aufgabe (2): Extrahieren Sie aus dem File 'opel-manta-listings.html' sämtliche Preisangaben der Fahrzeuge mit Hilfe eines geeigneten XPath.</b>
<b>Details zur Aufgabenstellung:</b>
- Sie finden das HTML-File im Order 'Data'.
- Komplettieren Sie für das Lösen der Aufgabe den unfertigen XPath im folgenden Python Code.
- Es muss nur dieser XPath angepasst werden, der restliche Code muss nicht modifiziert werden.
- Das Ergebnis (Object 'prices') sollte eine Liste in dieser Form sein: ["CHF 35'000", ... , "CHF 19'900"].

Tipp: Für das Testen des XPath können Sie diesen Online XPath-Tester verwenden: http://xpather.com

<b style="color: gray;">(max. erreichbare Punkte: 4)</b>

```python
# Import HTML-File
with open('./Data/opel-manta-listings.html', 'r', encoding='utf-8') as file:
    html_content = file.read()

# Parsen des HTML-Contents
tree = lxml_html.fromstring(html_content)

# XPath zum Extrahieren der Preisinformationen (bitte komplettieren)
try:
    prices = tree.xpath('/html/body/table/tbody/tr/td[2]/text()') # Hier den XPath ergänzen!
except:
    prices = [] # Ausgabe einer leeren Liste, falls der XPath nicht korrekt ist

# Anzeigen der extrahierten Preise
print("Extracted Prices:", prices)
```

#### <b>Aufgabe (4): Beziehen Sie die RSS-Feeds der deutschen Wochenzeitschrift 'DER SPIEGEL', und erstellen Sie aus den Titeln eine Word Cloud.</b>
<b>Details zur Aufgabenstellung:</b>
- Speichern Sie die RSS-Feeds in einem object 'feed' ab.
- Erstellen Sie eine Word Cloud Graphik auf der Basis der Titel der Artikel und stellen Sie diese im Jupyter Notebook dar.
- Stellen Sie die Titel der Artikel im Jupyter Notebook dar.

Tipp: Die URL für die aktuellen RSS-Feeds ist: https://www.spiegel.de/index.rss.

<b style="color: gray;">(max. erreichbare Punkte: 4)</b>

```python
# RSS-Feeds beziehen
feed = feedparser.parse("https://www.spiegel.de/index.rss")

# Daten aus dem RSS-Feed abrufen
feed_entries = feed.entries

# Leere Zeichenkette 'wordcloud_text' zur Speicherung des Textes initialisieren
wordcloud_text = ""

# Durch alle Einträge im Feed iterieren
for entry in feed.entries:

    # Artikel titel extrahieren
    article_title = entry.title

    # Titel zu 'wordcloud_text' hinzufügen
    wordcloud_text += f"{article_title} "

# Word Cloud generieren
wordcloud = WordCloud(width=700, 
                      height=700, 
                      background_color='black').generate(wordcloud_text)

# Word Cloud anzeigen
print("Word Cloud der Artikeltitel:")
plt.figure(figsize=(7, 7))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show() 

# Artikeltitel ausgeben
print("Artikeltitel:")
for entry in feed.entries:
    # Artikeltitel ausgeben
    print(f"{entry.title}")
```

#### <b>Aufgabe (5): Beziehen Sie mit Hilfe der Overpass Turbo Web API Daten zu den Sport-Stadien in der Schweiz und speichern Sie diese ab.</b>
<b>Details zur Aufgabenstellung:</b>
- Sport-Stadien sind mit dem key:value Paar "leisure"="stadium" in der OpenStreetMap Map-Feature Übersicht angegeben. Berücksichtigen Sie nur dieses key:value Paar.
- Speichern Sie die Daten in einem pandas DataFrame, und stellen Sie den DataFrame ganz oder in Teilen (z.B. erste 5 Zeilen) dar.
- Speichern Sie die Daten in einer SQLite Datenbank 'sports_stadiums.db' im Order 'Data'.
- Überprüfen Sie mit Hilfe von Python Code, ob die SQLite Datenbank 'sports_stadiums.db' generiert wurde.

Tipp: Auf der Webseite: https://overpass-turbo.eu können Sie Overpass Turbo Abfragen testen.

<b style="color: gray;">(max. erreichbare Punkte: 8)</b>

```python
# Overpass API URL
url = 'http://overpass-api.de/api/interpreter'

# Overpass turbo Abfrage-String
query = f"""
        [out:json];
        area["ISO3166-1"="CH"][admin_level=2];
        node ["leisure"="stadium"](area);
        out;"""

# Web API Anfrage
r = requests.get(url, params={'data': query})
data = r.json()['elements']

# Speichern der Daten in einem Pandas DataFrame
df_stadiums = json_normalize(data)

# Speichern der Daten in einer SQLite Datenbank
conn = sqlite3.connect('./Data/sports_stadiums.db')
df_stadiums.to_sql('stadiums', conn, if_exists='replace', index=False)
conn.close()

# Überprüfen, ob die SQLite Datenbank generiert wurde
file = [f for f in os.listdir('./Data') if re.match('sports_stadiums.db', f)]
print(file)

# Anzeigen der ersten 5 Zeilen des DataFrames
df_stadiums.head(5)
```
